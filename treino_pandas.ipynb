```python
import pandas as pd

df = pd.read_csv('./dataframe_treino_pandas.csv')
df
```

```python
df.info()
```

**Criando uma cópia do Dataframe**

```python
# Criando uma copia do Dataframe
df_copia = df.copy()
```

### Trantando coluna nome

```python
# Removendo elementos nulos na coluna nome
name_none_removed = df_copia['Nome'].fillna('Não Informado', inplace=True)

# Removendo espaços extras de cada elemento da coluna nome
nomes_sem_espacos = list(map(lambda n:n.strip(), df_copia['Nome']))

# passando os nomes para a coluna nome
df_copia['Nome'] = nomes_sem_espacos

df_copia
```

### Tratando coluna idade

```python
# Em caso de erro na conversão (p. ex., uma string que não representa um número), um valor vazio (null / nan) será
# atribuído no lugar
df_copia['Idade'] = pd.to_numeric(df_copia['Idade'], errors='coerce')
age_nan_removed = df_copia['Idade'].fillna(0, inplace=True)

# peguntar para a IA como deixar a idade no tipo inteiro e tirar do tipo float
df_copia
```

```python
# Convertendo a idade do tipo float para o tipo inteiro

df_copia['Idade'] = df_copia['Idade'].astype('Int64')

df_copia.loc[2,'Idade'] = 23
df_copia.loc[3,'Idade'] = 20

df_copia
```

### Tratando coluna cidade

```python

# Removendo dados nulos, Lembrando que o inplace = True serve para inplantar as alterações em definitivo
df_copia['Cidade'].fillna('Não Informado', inplace=True)

# Removendo espaços em valores da tabela
city_sem_espaco = list(map(lambda x:x.strip(), df_copia['Cidade']))

# Formantado as cidades para que comece com a letra maiúscula
city_capitalize = list(map(lambda x:x.capitalize(), city_sem_espaco))


# Colocando os valores limpos na coluna Cidade
df_copia['Cidade'] = city_capitalize

df_copia
```

### Tratando Coluna Salário

```python
df_copia
```

```python
# Aqui eu estou convertendo o salário para um valor numerico em float
# O coerce serve para colocar valores nulos quando houver algum erro de tipo na coluna
df_copia['Salario'] = pd.to_numeric(df['Salario'], errors='coerce')

# aqui eu estou resolvendo um erro de um valor que estava errado antes pois estava com vírgula e quando eu
# converti ele passou a ser nulo depois eu só consertei com a forma correta.
df_copia.loc[2, 'Salario'] = 4000.0
df_copia.loc[4, 'Salario'] = 5000.0
df_copia.loc[5, 'Salario'] = 5200.1


# Agora vou passar os dados nulos para 0
df_copia['Salario'].fillna(0, inplace=True)

df_copia

```

### Tratando a coluna Data_Admissao

```python
# Converte para datetime
df_copia['Data_Admissao'] = pd.to_datetime(df_copia['Data_Admissao'], errors='coerce')

# Atualiza valores específicos de forma segura
df_copia.loc[1, 'Data_Admissao'] = pd.Timestamp("2021-03-10")
df_copia.loc[2, 'Data_Admissao'] = pd.Timestamp("2020-04-15")
df_copia.loc[5, 'Data_Admissao'] = pd.Timestamp("2020-08-15")

# Preenche nulos com uma data padrão (datetime)
df_copia['Data_Admissao'].fillna(pd.Timestamp("1900-01-01"), inplace=True)

df_copia
```

### Tratando a coluna Departamento

```python
df_copia.loc[5,'Departamento']= "Marketing"
df_copia
```

### Tratando a coluna Horas_Trabalhadas

```python
#Convertendo as horas trabalhadas para numerico e se houver erro de tipo o errors='coerce' permite 
# colocar um valor nulo ao invés de dar erro de tipo
df_copia['Horas_Trabalhadas'] = pd.to_numeric(df_copia['Horas_Trabalhadas'], errors='coerce')

# Preenchendo todos os valores que não são numeros para o padrão 0 e inplace true para salvar as alterações como definitivas no df
df_copia['Horas_Trabalhadas'].fillna(0, inplace=True)

# Convertendo as horas para tipo inteiro
df_copia['Horas_Trabalhadas'] = df_copia['Horas_Trabalhadas'].astype('Int64')

# Corrigindo as horas trabalhadas das linhas que estão sem horas
df_copia.loc[2,'Horas_Trabalhadas'] = 170
df_copia.loc[4,'Horas_Trabalhadas'] = 160

# Apresentado novo dataframe
df_copia
```

# Salvando Dataframe e passando a cópia para o dataframe original

```python
#Passando as alterações feitas na cópia para o dataframe original
df = df_copia
df
```

```python
# Passando os dados tratados para um novo csv

df.to_csv("novo_Dataframe_treino_v2.csv")
```

**Selecionando funcionarios do departamento de TI**

```python
# selecionando apenas o departamento de TI
# OBS: Ao selecionar uma coluna string é necessário utilizar as aspas duplas ("") para definir
# qual departamento eu quero e a query precisa está entre aspas simples para funcionar
departamento_ti = df.query('Departamento == "TI"')
departamento_ti
```

**Selecionando quem ganha mais que 4000**

```python
# essa variavel armazena a selecao de pessoas de pessoas ganham mais que 4000
ganha_mais_4000 = df.query('Salario > 4000')
ganha_mais_4000
```

**Selecionando que tem menos de 30 anos**

```python
idade_maior_30 = df.query('Idade < 30')
idade_maior_30
```

**Criando coluna de salario anual**

```python
#Pegando cada item da coluna salario e multiplicando por 12 e ao mesmo tempo arredondando com a função round utilizando a função lambda
Salario_anual = list(map(lambda s:round(s*12, 2), df['Salario']))

df["Salario_Anual"] = Salario_anual
df
```

**Média Salarial**

```python
# Média Salarial

media_salarial = df['Salario'].mean()

print(media_salarial)
```

**Funcionarios por Departamento**

```python
# Agrupando com o groupby
# Depois contando com oa função  size() do pandas
# Depois resetando indice para ficar um dataframe organizado

grupo_departamento = df.groupby('Departamento').size().reset_index(name='Quantidade')
grupo_departamento
```

**Pagar o menor salário**

```python
# Pegando o valor minimo de salario
menor_salario = df['Salario'].min()
print(menor_salario)
```

**Horas Trabalhadas por cada departamento**

```python
# Agrupando departamentos e somando em seguida
gp_dept = df.groupby('Departamento')
gp_dept['Horas_Trabalhadas'].sum().reset_index(name="Soma_Horas_Departamento")
```
